# Import Libraries

```python
import bs4
from bs4 import BeautifulSoup
from pathlib import Path
import polars as pl

# Check Version
print(f"BeautifulSoup Version: {bs4.__version__}")
```

# Load test file

```python
html_file = Path("../data/source/alpha/alpha_week_2/02-2029_alpha_week_2.html") 

if html_file.exists():
    with open(html_file, "r", encoding="utf-8") as f:
        soup = BeautifulSoup(f, "lxml")
    
    for script_or_style in soup(["script", "style"]):
        script_or_style.decompose()

    raw_text = soup.get_text(separator="\n")
    lines = [line.strip() for line in raw_text.splitlines() if line.strip()]
    
    print(f"File loaded. {len(lines)} Rows found.")
else:
    print("File not found. Check the path.")
```

# Keywords and Results

```python
# Define search keywords
keywords = ["Gold", "Cost", "Attack Damage", "Ability Power", "Health", "Movement Speed"]

# flag rows with keywords and store them for analysis
found_patterns = []

for line in lines:
    active_flags = [k for k in keywords if k.lower() in line.lower()]
    
    if active_flags:
        found_patterns.append({
            "text": line,
            "flags": active_flags
        })

# Display the first 20 matches for analysis
for p in found_patterns[:20]:
    print(f"[{', '.join(p['flags'])}] -> {p['text']}")
```

# Copy raw files from source and remove pngs

```python
import shutil
from pathlib import Path

source_path = Path("../data/source")
bronze_path = Path("../data/bronze")

if bronze_path.exists():
    shutil.rmtree(bronze_path)

shutil.copytree(
    src=source_path, 
    dst=bronze_path, 
    ignore=shutil.ignore_patterns("*.png", "*.PNG")
)

html_count = len(list(bronze_path.rglob("*.html")))
png_count = len(list(bronze_path.rglob("*.png")))

print(f"Copying done.")
print(f"html in bronze: {html_count}")
print(f"png in bronze: {png_count}")
```
